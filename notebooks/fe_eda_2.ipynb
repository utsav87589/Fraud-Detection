### splitting the data

```python
df_categorical
```

```python
df_categorical.info()
```

### numeric columns

```python
df_numerical
```

```python
df_numerical.drop('Fraud_Label', axis = 1, inplace = True)
```

```python
df_numerical.info()
```

### saving the datasets

```python
df_categorical.to_csv('../data/categorical-only.csv')
df_numerical.to_csv('../data/numerical-only.csv')
```

### loading the numerical columns and applying the transformations

```python
df_numerical = pd.read_csv('../data/numerical-only.csv')
df_numerical.head()
```

```python
df_numerical.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
df_numerical
```

```python
df_numerical.info()
```

```python
### scaling will only be applied to the continuous columns like
### account balance, Avg_Transaction_Amount_7d, transaction_distance, and transaction_amount_boxcox
```

```python
df_numerical['Risk_Score'].min()
```

```python
df_numerical['Risk_Score'].max()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0, 1))
df_numerical['Account_Balance'] = scaler.fit_transform(df_numerical[['Account_Balance']])
```

```python
df_numerical.head()
```

```python
df_numerical['Avg_Transaction_Amount_7d'] = scaler.fit_transform(df_numerical[['Avg_Transaction_Amount_7d']])
```

```python
df_numerical['Transaction_Distance'] = scaler.fit_transform(df_numerical[['Transaction_Distance']])
```

```python
df_numerical['Transaction_Amount_boxcox'] = scaler.fit_transform(df_numerical[['Transaction_Amount_boxcox']])
```

```python
df_numerical.head()
```

```python
df_numerical['Card_Age'].value_counts()
```

### saving the feature for the distance based algos

```python
df_numerical.to_csv('../data/numerical_for_distance.csv')
```

### making the final changes for the algos

```python
df_categorical = pd.read_csv('../data/categorical-only.csv')
df_categorical.drop('Unnamed: 0', axis = 1, inplace = True)
df_categorical.head()
```

```python
df_numerical = pd.read_csv('../data/numerical-only.csv')
df_numerical.drop('Unnamed: 0', axis = 1, inplace = True)
df_numerical.head()
```
